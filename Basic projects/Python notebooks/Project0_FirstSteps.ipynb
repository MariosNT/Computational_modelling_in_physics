### 0.3.3 Finding Roots of Quadratic Function using Bisection Method

```python
def bisection(xl, xr, eps, func):
    """
    Finds and returns a root of the given function within the specified
    tolerance using the bisection method.

    Parameters
    ----------
    xl : float
        The lower bound of the search interval.
    xr : float
        The upper bound of the search interval.
    eps : float
        The tolerance within which the root is found.
    func : function
        A Python function representing the equation whose root is to be found.
        This function should accept a single argument and return a float.

    Returns
    -------
    float
        An approximation to the root of the equation represented by the
        input function within the specified tolerance.
    """
    while xr - xl > eps:
        s = (xl + xr) / 2
        if func(s) == 0:
            return s
        elif func(xl) * func(s) < 0:
            xr = s
        else:
            xl = s
    return (xl + xr) / 2


def get_zero_points(xl, xr, eps, func):
    """
    Returns a list of the zero points of the specified function with epsilon
    precision.

    Parameters
    ----------
    xl : float
        The left limit of the range in which to search for the zero points.
    xr : float
        The right limit of the range in which to search for the zero points.
    eps : float
        The precision of the results.
    func : function
        The function whose values are computed.

    Returns
    -------
    list
        A list of the zero points of the specified function with epsilon
        precision.
    """
    return [bisection(xl, (xl + xr) / 2, eps, func),
            bisection((xl + xr) / 2, xr, eps, func)]


def get_function_values(xl, xr, func):
    """
    Returns a dictionary with two numpy arrays: x-points and y-values of the
    given function.

    Parameters
    ----------
    xl : float
        The left limit of the range for which the function values are
        calculated.
    xr : float
        The right limit of the range for which the function values are
        calculated.
    func : function
        The function whose values are computed.

    Returns
    -------
    dict
        A dictionary with two keys, 'x' and 'y', containing numpy arrays
        of x-points and y-values.
    """
    values = {'x':[], 'y':[]}
    while xl < xr:
        values['x'].append(xl)
        values['y'].append(func(xl))
        xl += 0.01
    return values

def square_function(x):
    """
    Computes the value of a square function of the form
    f(x) = 4x^2 + 2x - 6 for a given input value x.

    Parameters
    ----------
    x : float
        The input value for the square function.

    Returns
    -------
    float
        The computed value of the square function for the input x.
    """
    return 4 * x ** 2 + 2 * x - 6

# defining the necessary variables and calculating the zero points
# of the example function
x_left, x_right = -3.0, 3.0
epsilon = 1e-15
zero_points = get_zero_points(x_left, x_right, epsilon, square_function)
print("Zero points of the function: ", zero_points)

# visualization of collected data
function_values = get_function_values(x_left, x_right, square_function)
fig3 = plt.figure(figsize=(8, 5))
plt.plot(function_values['x'], function_values['y'], label='$4x^2$ + 2x - 6')
plt.plot(function_values['x'], [0 for x in function_values['x']],
         linestyle='dashed', linewidth=0.8, color='gray')
plt.plot(zero_points, [0 for x in zero_points], 'ro')
plt.legend(fontsize=10)
plt.show()
```

### 0.3.1 Finding the Minimum of 1D Function using Golden Section Method

```python
def simple_square_funtion(x):
    """
    Returns the value of the square function at the point x.

    Parameters
    ----------
    x : float
        The point at which the square function value is computed.

    Returns
    -------
    float
        The computed value of the square function at the point x.
    """
    return x * x - 1

def cosinus_function(x):
    """
    Returns the value of the cosine function at the point x.

    Parameters
    ----------
    x : float
        The point at which the cosine function value is computed.

    Returns
    -------
    float
        The value of the cosine function at the point x.
    """
    return np.cos(x)

def golden_section_search(a, b, eps, func):
    """
    Finds the minimum of a given function using golden section search.

    Parameters
    ----------
    a : float
        The left endpoint of the interval to search for the minimum.
    b : float
        The right endpoint of the interval to search for the minimum.
    eps : float
        The precision of the search. The algorithm will stop when the length
        of the interval is less than eps.
    func : function
        The function to minimize.

    Returns
    -------
    float
        The point at which the function reaches its minimum.
    """
    k = (np.sqrt(5) - 1) / 2
    
    # Define the two middle points
    # Always xl<xr (with the definition of k)
    xl = b - k * (b - a)
    xr = a + k * (b - a)
    
    # Move all the points with the golden-ratio step,
    # closing down to the minimum
    while (b - a) > eps:
        if(func(xl) < func(xr)):
            b = xr
            xr = xl
            xl = b - k * (b - a)
        else:
            a = xl
            xl = xr
            xr = a + k * (b - a)
    return (a + b) / 2



def plot_minimum(func_val, label, minimum, func):
    """
    Plots the given function and its minimum.

    Parameters
    ----------
    func_val : dict
        The dictionary containing two lists. The x-points list and the list of
        values at those points of the given function.
    label : str
        The label for the plot of the function.
    minimum : float
        The point at which the function reaches its minimum.
    func : function
        The function to plot.
    """
    plt.rcParams['legend.fontsize'] = 10
    plt.plot(func_val['x'], func_val['y'], label=label)
    plt.plot(func_val['x'], [0 for x in func_val['x']], linestyle='dashed',
             linewidth=0.8, color='gray')
    plt.plot(minimum, func(minimum), 'ro')
    plt.legend()
    plt.show()
```

Examples of Applications:
   - minimum of the quadratic function f(x) = $x^2$ - 1 (expected x = 0)
   - minimum of the function f(x) = cos(x) (expected π)

```python
epsilon_3 = 1e-15

minimum_1 = golden_section_search(-1.0, 1.0, epsilon_3, simple_square_funtion)
print('Minimum of the function (x^2 - 1) determined using the golden section method:\n ',
      minimum_1)
function_values_1 = get_function_values(-1, 1, simple_square_funtion)
plot_minimum(function_values_1, '$x^2$ - 1', minimum_1, simple_square_funtion)

minimum_2 = golden_section_search(1.5, 5.0, epsilon_3, cosinus_function)
print('\nMinimum of the function cos(x) determined using the golden section method:\n ',
      minimum_2, '[approximation of the number PI]')
function_values_2 = get_function_values(1.5, 5, cosinus_function)
plot_minimum(function_values_2, 'cos(x)', minimum_2, cosinus_function)
```
